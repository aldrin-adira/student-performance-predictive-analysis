### Dataset Overview
The dataset consists of 2,100 student records across 7 features:
- **Personal Detail:** Student ID, Age
- **Study Habits & Parental Involvement:** Weekly Study Hours, Absences, Parental Support Level
- **Academic Performance:** GPA, Grade Class

## Data Cleaning and Quality Control

```python
student_dataset['GradeClass'].dtypes
```

```python
student_dataset['GradeClass'] = student_dataset['GradeClass'].replace(
{0:'A', 1:'B', 2:'C', 3:'D', 4:'F'}
)

student_dataset['GradeClass'].dtypes
```

The code above has replaced the numerical representation of the grade class to an alphabetical representation. This change also results in the change of the data type from being an integer to an object.
